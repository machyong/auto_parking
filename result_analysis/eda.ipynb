```python
score = pd.read_csv('/home/yong/auto_parking/saved_model/episode_score0827.csv', header = None)
result = pd.read_csv('/home/yong/auto_parking/saved_model/episode_results0827.csv', header = None)
result.drop_duplicates(0, inplace= True)
score2 = score.drop_duplicates(0).reset_index()
result2 = result.reset_index()
```

```python
score2['result'] = result2[1]
score2.columns = ['index','episode','total_step','score','result']
score2.drop(columns='index', inplace=True)
```

```python
score3 = score2[score2['result'] == 'succeed']
score3.head()
```

```python
def func1(x):
    if x <= 100:
        return 0
    elif x <= 200:
        return 100
    elif x <= 300:
        return 200
    elif x <= 400:
        return 300
    elif x <= 500:
        return 400
    elif x <= 600:
        return 500
    elif x <= 700:
        return 600
    elif x <= 800:
        return 900
    elif x <= 900:
        return 800
    else:
        return 900
    
score3['episode_category'] = score3['episode'].apply(lambda x: func1(x))
score3['episode_category'].value_counts()
```

```python
score3.info()
```

```python
sns.scatterplot(data=score3,x='episode',y='score', hue='episode_category', palette='rainbow')
plt.show()
```

```python
sns.scatterplot(data=score3,x='total_step',y='score', hue='episode_category', palette='rainbow')
plt.show()
```
